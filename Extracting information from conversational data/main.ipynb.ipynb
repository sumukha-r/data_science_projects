```python
#Using the with open method to open text file and reading it using readlines function which returns a list

with open("conv.txt") as file:
    data = file.readlines()
    
#The text file has special characters hence decoding using utf-8
#Stripping to remove "\n" 
#Splitting the line after the semicolon which denotes the separation of speaker name and the dialogue
#Adding only lines which have actual dialogues and not the blank ones
#Flatmapping the list

conv = [i.encode("windows-1252").decode("utf-8").strip().split(":") for i in data if (i != "\n")]
conv = [i for j in conv for i in j]
conv
```

```python
#Creating a set to store names of all the speakers in the conversation

import string

speakers = set()
for i in conv:
    
    #Using upper method as all names are in uppercase
    #Using punctuation to make sure other words such as "NO!" are not included
    
    if i.isupper() & all(list(j not in string.punctuation for j in i)):
        speakers.add(i)
print(f"The speakers are: {list(speakers)} and the total number of speakers are {len(speakers)}")
```

```python
#Creating a dictionary to store the speaker name and their respective dialogues

all_dialogues = {}
for i in speakers:
    dialogues = []
    for (j, k) in zip(conv, [*range(len(conv))]):
        
        #As split is used, the dialogue of the speaker will be the next item in the list
        
        if j.startswith(i):
            dialogues.append(conv[k+1])
    all_dialogues[i] = dialogues
all_dialogues
```

```python
#Creating a dictionary to store only the unique words spoken by the speaker

import re

unique_words = {}
for speaker in all_dialogues.keys():
    words = set()
    all_words = []
    for i in all_dialogues[speaker]:
        i = i.strip().split()
        all_words.append(i)
    all_words = [j for k in all_words for j in k]
    
    #Using regex module to filter out punctuation marks present in the words
    
    all_words = [re.sub(r'[^\w\s]', "", i) for i in all_words if i != "..."]
    all_words = list(filter(None, all_words))
    
    #Using a set to make use there are no repitions of words 
    #Making all words lowercase to filter out even more (The and the)
    
    for word in all_words:
        words.add(word.lower())
    unique_words[speaker] = list(words)
unique_words
```

```python
#Writing unique words of all the speaker in each file where each word is in a new line

for i in unique_words.keys():
    with open(f"{i}.txt", "w+") as file:
        file.writelines("\n".join(unique_words[i]))
```
